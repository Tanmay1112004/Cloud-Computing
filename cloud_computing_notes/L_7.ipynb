# üìò **UNIVERSITY NOTES (Simple Language + Exam-Friendly)**

## **1. Cloud Architecture Hierarchy**

### **1.1 Region**

A *region* is a **large geographical area** where a cloud provider hosts its services.
Example: "Central India", "East US".

### **1.2 Availability Zone (AZ)**

An AZ is a **physically separate location inside a region**, with its own power, cooling, and networking.
If one zone fails, the other zones continue working.

### **1.3 Data Center**

A data center is the **actual building** inside an AZ that contains:

* servers
* storage devices
* networking equipment
* cooling systems

Multiple data centers together form an availability zone.

---

## **2. Cloud Resources**

Cloud resources are the **core components** provided by cloud platforms.
The three most important resources (as per the video) are:

1. **Storage** ‚Üí Stores data securely.
2. **Networking** ‚Üí Connects systems and routes data.
3. **Compute** ‚Üí Provides processing power (VMs, containers).

---

## **3. Latency (Simple Explanation)**

**Latency** is the *time delay* between sending a request and receiving a response.
Less distance = less latency.
This is why cloud providers keep data centers close to users.

Example:
If you're in Mumbai and the server is in Japan, the delay will be higher.

---

## **4. Fault Tolerance**

**Fault tolerance** means the system keeps working even if some part fails.
Cloud uses:

* multiple availability zones
* multiple data centers
  to make sure apps never go down.

---

## **5. Why Use Availability Zones?**

* Prevent downtime
* Avoid single point of failure
* Improve high availability
* Better disaster recovery

If one AZ fails (power cut, flood, fire), workloads shift to other zones.

---

## **6. Why Cloud Uses Hierarchical Structure (Main Reason)**

* Better performance
* Low latency
* Scalability
* High reliability
* Disaster recovery

# üèô **Analogy: Region, Zone, Data Center**

Think of a big company:

* **Region = Country**
* **Availability Zone = City inside that country**
* **Data Center = Office building inside that city**

If one building shuts down, work continues from another building or city.

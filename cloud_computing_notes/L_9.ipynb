# **1. Introduction to Cloud Computing**

**Definition:**
Cloud computing is the delivery of computing services ‚Äî such as servers, storage, databases, networking, software ‚Äî over the internet (‚Äúthe cloud‚Äù).
Instead of owning hardware, users *rent* resources based on their needs.

**Characteristics:**

* On-demand access
* Pay-as-you-go
* Scalability
* No physical maintenance

---

# **2. Cloud Service Models**

Cloud services are divided into three main types:

* **IaaS ‚Äì Infrastructure as a Service**
* **PaaS ‚Äì Platform as a Service**
* **SaaS ‚Äì Software as a Service**

---

# **3. Infrastructure as a Service (IaaS)**

### **Definition:**

IaaS provides virtualized computing infrastructure over the internet. Users have maximum control.

### **What the User Manages:**

* Operating System
* Middleware
* Runtime
* Applications
* Data

### **What the Provider Manages:**

* Virtualization
* Servers
* Storage
* Networking

### **Analogy:**

Renting a kitchen with an oven and electricity. You bring ingredients + recipe.

### **Examples:**

* AWS EC2
* Microsoft Azure VM
* Google Compute Engine

### **Use Cases:**

* Custom OS requirements
* Hosting large apps
* Migration from on-premise systems
* Full control environments

---

# **4. Platform as a Service (PaaS)**

### **Definition:**

PaaS provides a ready-made development platform, tools, and environment.

### **What the User Manages:**

* Applications
* Data

### **What the Provider Manages:**

* Runtime
* Middleware
* OS
* Virtualization
* Servers
* Networking

### **Analogy:**

A fully-equipped kitchen where you only bring the recipe.

### **Examples:**

* Google App Engine
* Heroku
* AWS Elastic Beanstalk
* Azure App Service

### **Use Cases:**

* App development
* Rapid prototyping
* Developers who don‚Äôt want to manage infrastructure

### **Main Benefit:**

You focus on coding; all infrastructure headaches vanish.

---

# **5. Software as a Service (SaaS)**

### **Definition:**

SaaS delivers ready-to-use applications over the internet.

### **What the User Manages:**

* Nothing (only uses the software)

### **What the Provider Manages:**

* Entire application stack
* Infrastructure
* Updates
* Security

### **Analogy:**

Pizza delivery ‚Äî just eat.

### **Examples:**

* Netflix
* Gmail
* Google Docs
* Salesforce
* Canva

### **Use Cases:**

* Email
* Online office tools
* CRM systems
* Streaming apps

---

# **8. University-Type Questions & Answers**

### **Q1. Differentiate between IaaS, PaaS, and SaaS. Give one example each.**

**Answer:**

* **IaaS:** Cloud provider manages infrastructure. User manages OS, apps, data.
  *Example:* AWS EC2
* **PaaS:** Cloud provider manages infrastructure + platform. User manages apps & data.
  *Example:* Google App Engine
* **SaaS:** Provider manages everything; user simply uses the software.
  *Example:* Netflix

---

### **Q2. Why is PaaS suitable for startups?**

**Answer:**
PaaS removes the need to manage servers, OS, runtime, and hardware. Developers can focus only on writing and deploying code, making it ideal for startups with limited technical staff.

---

### **Q3. A user accesses a software through a browser without installation. Which model is this?**

**Answer:**
**SaaS**, because the entire application is hosted by the provider and used via the internet.

---

### **Q4. What is the main benefit of PaaS?**

**Answer:**
It allows developers to build and deploy applications quickly without worrying about infrastructure management.

---

### **Q5. Explain data security responsibilities across service models.**

**Answer:**

* **SaaS:** Provider handles most security.
* **PaaS:** Shared responsibilities.
* **IaaS:** User handles majority of security including OS-level patches.

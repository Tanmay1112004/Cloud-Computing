## **1. Introduction to Cloud Computing**

### **What is Cloud Computing?**

Cloud computing means using **internet-based resources** like servers, storage, networking, and databases instead of buying physical hardware.

### **Why is Cloud Important?**

* Cloud is growing fast because **AI and ML need heavy hardware**, which cloud provides.
* More companies ‚Üí more jobs in cloud.
* Scalable: add/remove resources anytime.

### **Career Opportunities**

* Cloud Engineer
* Cloud Architect
* Cloud Administrator
* DevOps Engineer
* AI/ML Engineer

# **üìù UNIVERSITY-STYLE QUESTIONS & ANSWERS**

### **Q1. What is cloud computing?**

Cloud computing is the delivery of computing services‚Äîsuch as storage, servers, networking, and databases‚Äîover the internet, allowing users to access resources without owning physical hardware.

---

### **Q2. Why is cloud computing important in AI and Machine Learning?**

AI and ML require powerful GPUs and high-performance servers. Cloud providers offer these resources on demand, making it easier and cheaper to train and deploy AI models.

---

### **Q3. Explain virtualization in cloud computing.**

Virtualization allows one physical machine to run multiple virtual machines (VMs). This helps cloud providers efficiently share hardware and offer scalable resources to users.

---

### **Q4. What is EC2?**

EC2 (Elastic Compute Cloud) is an AWS service that provides virtual servers in the cloud. You can use EC2 to run applications, host websites, or perform heavy computing tasks.

---

### **Q5. List any three cloud service models.**

1. IaaS ‚Äì Infrastructure as a Service
2. PaaS ‚Äì Platform as a Service
3. SaaS ‚Äì Software as a Service

---

### **Q6. What are the benefits of cloud computing?**

* Scalability
* Cost-effective
* Reliable
* Global accessibility
* No hardware maintenance

---

### **Q7. How do certifications validate cloud knowledge?**

Certifications prove you understand cloud concepts, services, architecture, and best practices through standardized exams recognized by the industry.

---

### **Q8. What is the purpose of certifications (one-line answer)?**

To officially verify and prove your technical skills to employers.

# **2. Hosting Models Explained (with Car Analogy)**

## **A. On-premises Hosting**

**Analogy:** *Buying your own car.*
You buy it, maintain it, repair it, and control everything.

### **Definition:**

Organization owns and manages all hardware, software, and data inside its own building.

### **Key Features:**

* **Ownership:** Full
* **Control:** Complete
* **Setup Time:** Very long (buying hardware, installation)
* **Cost:** Very high upfront
* **Scalability:** Low (upgrading takes time/money)
* **Security:** Highest (good for sensitive data)

### **Used By:**

Banks, financial companies, governmentâ€”because they need full control.

## **B. Dedicated Hosting**

**Analogy:** *Renting a car monthly.*
You get the car only for yourself, but you donâ€™t own it.

### **Definition:**

A company rents a physical server from a provider.
The hardware is owned by the provider, but you get full dedicated use.

### **Key Features:**

* **Ownership:** Rented
* **Control:** High
* **Setup Time:** Moderate
* **Cost:** Monthly rental
* **Scalability:** Moderate (still slower than cloud)
* **Security:** Good but depends on provider

### **Used By:**

Companies that need more control but donâ€™t want to buy hardware.

---

## **C. Cloud Computing**

**Analogy:** *Using Uber/Ola daily.*
You donâ€™t own the car, you just pay whenever you use it.

### **Definition:**

Servers, storage, and services are provided online by a cloud provider.
You only pay for what you use.

### **Key Features:**

* **Ownership:** Shared
* **Control:** Medium
* **Setup Time:** Instant
* **Cost:** Pay-as-you-use (metered service)
* **Scalability:** Very high (add/remove instantly)
* **Security:** Shared responsibility

### **Used By:**

Startups, growing companies, apps with changing workloads.

# **8. Short 6-Mark Answer (Perfect for Exams)**

**Explain Cloud Computing and why it is better than traditional hosting.**

**Answer:**
Cloud computing provides IT resources like servers, storage, and software over the internet on a pay-as-you-use model. It requires no upfront cost, supports instant setup, and offers very high scalability. Compared to traditional hosting like on-premises or dedicated hosting, cloud computing is faster to deploy, cheaper for most users, easier to scale, and reduces the need to manage hardware. It is ideal for startups, growing companies, and applications with fluctuating demand.

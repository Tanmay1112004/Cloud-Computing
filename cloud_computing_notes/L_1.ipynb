## **3. On-Premises vs Hosted vs Cloud**

### **A) On-Premises (Traditional IT)**

* You **buy** hardware & software.
* You manage **everything** (maintenance, updates, security).
* **High upfront cost**.
* **Low flexibility** ‚Üí scaling requires buying new equipment.

**Example:**
Buying your own laptop.

---

### **B) Hosted Server**

* You **rent hardware** from a vendor for a fixed time.
* Vendor manages hardware; you manage software.
* **Fixed payment**, even if resources stay unused.
* Limited flexibility.

**Example:**
Renting a laptop for 6 months even if you need it for only 1 month.

---

### **C) Cloud Computing**

* You use resources available on the internet.
* **Pay only for what you use** (metered service).
* No upfront cost.
* Highly **scalable** and flexible.
* Provider handles infrastructure.

**Example:**
Paying for laptop usage per hour, like electricity.

---

## **4. Key Characteristics of Cloud Computing**

### ‚úî **1. Pay-as-you-use (Metered Service)**

You pay only for the resources consumed.

### ‚úî **2. On-demand Self Service**

You can start or stop resources anytime.

### ‚úî **3. Scalability & Elasticity**

Resources increase or decrease automatically based on demand.

### ‚úî **4. No Upfront Cost**

No need to buy hardware/software.

### ‚úî **5. Multi-tenant Environment**

Many customers share the same infrastructure.

### ‚úî **6. Virtualization**

Multiple virtual machines run on one physical server.

### ‚úî **7. High Availability**

Services are available 24/7.

### ‚úî **8. Fast Deployment**

New servers or services can be created in seconds.

---

## **5. Cloud as a Large Data Center**

Cloud providers run huge data centers with thousands of interconnected servers.

These servers:

* divide tasks,
* run operations in parallel,
* hide the complexity from users.

Users only see the final result ‚Äî not the hardware behind it.

### **Q1. Define Cloud Computing.**

**Ans:**
Cloud computing is the on-demand delivery of computing resources (such as servers, storage, databases, and software) over the internet, where users pay only for the resources they use. It removes the need for owning or managing physical infrastructure.

---

### **Q2. Explain the difference between On-Premises and Cloud Computing.**

**Ans:**

| Feature        | On-Premises             | Cloud                           |
| -------------- | ----------------------- | ------------------------------- |
| Ownership      | User owns all hardware  | Provider owns hardware          |
| Cost           | High upfront cost       | Pay-as-you-use                  |
| Control        | Full control            | Limited control                 |
| Scalability    | Low                     | High & instant                  |
| Responsibility | User manages everything | Provider manages infrastructure |

**Analogy:**

* On-premises ‚Üí owning a car
* Cloud ‚Üí using Uber

---

### **Q3. What is ‚ÄúMetered Service‚Äù in cloud computing?**

**Ans:**
Metered service means cloud usage is billed according to actual consumption, similar to electricity. You only pay for the compute time, storage, or bandwidth you use. No usage = no cost.

### **Q5. Why is cloud computing called ‚Äúold wine in a new bottle‚Äù?**

**Ans:**

### **Q6. What technologies enable cloud computing?**

**Ans:**

2. **Distributed/Parallel Computing** ‚Äì divides tasks among multiple servers.

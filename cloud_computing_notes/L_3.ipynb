# **2. WHAT IS CLOUD COMPUTING? (Simple Definition)**

Cloud computing means **using computing services over the internet instead of owning physical hardware**.

Examples of cloud services:

* Storing files online ‚Üí Google Drive
* Streaming movies ‚Üí Netflix
* Streaming music ‚Üí Spotify

You don‚Äôt own the servers. You just *use* them.

---

# **3. WHY CLOUD COMPUTING?**

Using the e-bike business example:

* Buying servers = expensive
* Maintaining infrastructure = need skilled IT staff
* Scaling to more customers = difficult

Cloud solves all this by letting you rent IT services instantly.

# **5. KEY CHARACTERISTICS OF CLOUD COMPUTING**

### **1) On-Demand Self-Service**

You can get resources like storage or servers **instantly**, without calling anyone.
Example: buying extra iCloud storage.

---

### **2) Pay-as-You-Go (Metered Billing)**

You pay *only for the amount you use*.
Like Uber: pay for distance, not ownership.

---

### **3) Broad Network Access**

Access cloud services from **anywhere**, using any device (phone, laptop, tablet).

---

### **4) Resource Pooling**

Resources (like servers, storage) are **shared** among many users, but allocated logically.

**Analogy (exam-friendly):**
A hotel swimming pool ‚Üí many guests share it, but everyone gets access.

---

### **5) Scalability / Elasticity**

You can increase or decrease resources automatically based on demand.

**Example:**
If your app suddenly gets 1,00,000 users, cloud instantly gives more power‚Äîno new hardware needed.

**True/False:**
‚ÄúScalability means buying new hardware.‚Äù ‚Üí **False**

# **6. WHAT ARE CLOUD SERVICES? (IaaS, PaaS, SaaS)**

### **1) IaaS ‚Äì Infrastructure as a Service**

Provides virtual machines, storage, servers.
Example: AWS EC2

### **2) PaaS ‚Äì Platform as a Service**

Provides a platform for developers to build and deploy apps.
Example: Google App Engine

### **3) SaaS ‚Äì Software as a Service**

Software delivered over the internet.
Example: Netflix, Gmail

---

# **7. CLOUD DEPLOYMENT MODELS**

### **1) Public Cloud**

Owned by cloud providers, shared by many users.
Example: AWS, Azure

### **2) Private Cloud**

Used by one single organization.
More secure, more expensive.

### **3) Hybrid Cloud**

Mix of public + private cloud.
Used by companies needing flexibility and security.

---

# **8. WHAT IS VIRTUALIZATION?**

Virtualization is the technology that allows **one physical machine to run multiple virtual machines**.

It enables:

* Resource pooling
* Better efficiency
* Scalability
* Reduced costs

Example:
A single physical server ‚Üí runs 10 virtual servers for different customers.

* **Cost-effective** ‚Üí No need to buy hardware
* **Easy to manage** ‚Üí Provider handles maintenance
* **Highly scalable** ‚Üí Add/remove resources instantly
* **Accessible from anywhere**
* **Pay only for usage**

* **No technical setup needed**

# **üìå Answers to the Questions**

### **1. What is the difference between IaaS and SaaS?**

**IaaS (Infrastructure as a Service):**
Provides virtual hardware like servers, storage, and networking.
You manage the OS, apps, and data.
**Example:** AWS EC2.

**SaaS (Software as a Service):**
Provides ready-to-use software over the internet.
You don‚Äôt manage anything ‚Äî just use the app.
**Example:** Gmail, Netflix.

**Difference in one line:**
*IaaS = rent hardware; SaaS = use finished software.*

---

### **2. Which characteristic of cloud computing allows users to provision services instantly?**

**On-Demand Self-Service**
It lets users get resources (storage, servers, etc.) instantly without human interaction.

---

### **3. True or False: Cloud computing removes the need for IT maintenance.**

**True (for the customer).**
The cloud provider handles hardware, updates, security patches, etc., so the user doesn‚Äôt need to maintain physical infrastructure.

* **Netflix**

(You can also use Gmail, Spotify, Canva, Zoom ‚Äî all SaaS.)

**Scalability allows resources to increase or decrease automatically based on demand, ensuring the system performs smoothly even during high traffic, without buying new hardware.**

In one line:
*Your app never crashes during heavy traffic.*

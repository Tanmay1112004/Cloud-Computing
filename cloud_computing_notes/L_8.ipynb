# **Unit: Cloud Computing Service Models**

---

# ‚≠ê 1. **What is Cloud Computing?**

Cloud computing means accessing computing services (storage, servers, databases, applications) over the internet instead of owning them physically.
You pay only for what you use.

---

# ‚≠ê 2. **On-Premises (Traditional Model)**

You manage everything yourself.

### **You manage:**

* Applications
* Data
* Runtime
* Operating System
* Virtualization
* Servers
* Storage
* Networking

### **Analogy:**

You make pizza at home‚Äîbuy ingredients, cook it, clean everything.

---

# ‚≠ê 3. **IaaS (Infrastructure as a Service)**

The cloud provider gives the **infrastructure**, and *you manage the software layer*.

### **Cloud provider manages:**

* Virtualization
* Servers
* Storage
* Networking

### **You manage:**

* OS
* Runtime
* Applications
* Data

### **Analogy:**

You cook the pizza, but you use someone else‚Äôs kitchen (infrastructure).
You still do most of the cooking work.

### **Examples:**

* Amazon EC2
* Google Compute Engine
* Azure Virtual Machines

### **Use Cases:**

* Hosting websites
* Running custom apps
* Migrating old systems to cloud

---

# ‚≠ê 4. **PaaS (Platform as a Service)**

Cloud gives the full **platform** for development and deployment.

### **Cloud provider manages:**

* OS
* Runtime
* Middleware
* Servers, Storage, Networking

### **You manage:**

* Applications
* Data

### **Analogy:**

You bring the toppings, but the pizza base, oven, and chef are already provided.

### **Examples:**

* Google App Engine
* AWS Elastic Beanstalk
* Azure App Service

### **Use Cases:**

* Web app development
* Fast prototyping
* Deploying apps without maintaining infrastructure

---

# ‚≠ê 5. **SaaS (Software as a Service)**

Ready-to-use applications delivered over the internet.

### **Cloud provider manages EVERYTHING.**

You just use the software.

### **Analogy:**

Ordering a pizza. It arrives ready. You just eat.

### **Examples:**

* Gmail
* Google Docs
* Dropbox
* Salesforce
* Microsoft 365

### **Use Cases:**

* Email
* CRM
* File storage
* Collaboration tools

---

# ‚≠ê 6. **Key Differences (Short Summary)**

| Model           | Who Manages Most?     | User Control | Best For             |
| --------------- | --------------------- | ------------ | -------------------- |
| **On-Premises** | User                  | Highest      | Full control setups  |
| **IaaS**        | Shared                | High         | Custom apps, servers |
| **PaaS**        | Provider              | Medium       | Developers           |
| **SaaS**        | Provider (Everything) | Low          | Ready-to-use apps    |

# ‚≠ê 8. **Cloud Deployment Models**

These tell WHERE the cloud is deployed.

### **1. Public Cloud**

* Services shared over the internet for anyone
* Example: AWS, Azure
* Cheapest
* Best for startups, students, new apps

### **2. Private Cloud**

* Owned by one organization
* Highly secure
* Expensive
* Best for banks, government

### **3. Hybrid Cloud**

* Mix of public + private
* Best for companies needing both flexibility and security

### **4. Community Cloud**

* Shared by organizations with similar needs
* Example: hospitals, universities

---

# ‚≠ê 9. **Extra Explanation: IaaS (Deep but Simple)**

IaaS gives you **virtual hardware** in the cloud.

You get:

* Virtual machines
* Storage
* Networks
* Load balancers

You DO NOT get:

* OS management
* Application management

### Why IaaS?

* Highest flexibility
* You can install whatever OS you want
* You can scale servers up/down anytime
* Best when migrating old apps to cloud
